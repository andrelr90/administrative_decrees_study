## Imports

```python
import pandas as pd
import numpy as np
from sklearn.metrics import classification_report
```

```python
svm_test_prediction_sen = pd.read_csv("test_results/SVM_sen.csv", header=None).values
cnn_test_prediction_sen = pd.read_csv("test_results/CNN_sen.csv", header=None).values
han_test_prediction_sen = pd.read_csv("test_results/HAN_sen.csv", header=None).values

svm_test_prediction_minist = pd.read_csv("test_results/SVM_minist.csv", header=None).values
cnn_test_prediction_minist = pd.read_csv("test_results/CNN_minist.csv", header=None).values
han_test_prediction_minist = pd.read_csv("test_results/HAN_minist.csv", header=None).values
```

```python
def string_to_array_serie(serie):
    return np.array([np.array([int(i) for i in t.replace("[", "").replace("]", "").split()]) for t in serie])

test_set = pd.read_csv('sets/Set 0.csv')

Teste_Y_sen = string_to_array_serie(test_set["Senado"])
Teste_Y_minist = string_to_array_serie(test_set["Referenda"])
```

## Voting test function

```python
def voting_test(svm_test_prediction, cnn_test_prediction, han_test_prediction, threshold):
    votings = []
    for i in range(len(svm_test_prediction)):
        voting = svm_test_prediction[i]
        voting = np.add(voting, cnn_test_prediction[i])
        voting = np.add(voting, han_test_prediction[i])
        votings.append(voting)

    for i in range(len(votings)):
        for j in range(len(votings[i])):
            votings[i][j] = 1 if (votings[i][j] >= threshold) else 0
    votings = np.array(votings)
    return votings
```

## Custom metrics

```python
def acerto_absoluto(test_y_sen, test_pred, number_labels):
    acerto_100 = 0
    test_prediction_binary = test_pred
    for i in range(len(test_prediction_binary)):
        acertos_internos = 0
        for j in range(len(test_prediction_binary[i])):
            if test_y_sen[i][j] == test_prediction_binary[i][j]:
                acertos_internos+=1
        if acertos_internos == number_labels:
            acerto_100+=1
    return acerto_100/len(test_prediction_binary)*100
```

## Results Senate

```python
print("Senate voting 1 results")
votings = voting_test(svm_test_prediction_sen, cnn_test_prediction_sen, han_test_prediction_sen, 1)
print(classification_report(y_true=np.array(Teste_Y_sen), y_pred=votings, digits=3, ))
for i in range (len(Teste_Y_sen[0])+1):
    print(str(i)+"/"+str(len(Teste_Y_sen[0]))+" classes:", acerto_absoluto(Teste_Y_sen, votings, i))

print("\n\nSenate voting 2 results")
votings = voting_test(svm_test_prediction_sen, cnn_test_prediction_sen, han_test_prediction_sen, 2)
print(classification_report(y_true=np.array(Teste_Y_sen), y_pred=votings, digits=3, ))
for i in range (len(Teste_Y_sen[0])+1):
    print(str(i)+"/"+str(len(Teste_Y_sen[0]))+" classes:", acerto_absoluto(Teste_Y_sen, votings, i))
    
print("\n\nSenate voting 3 results")
votings = voting_test(svm_test_prediction_sen, cnn_test_prediction_sen, han_test_prediction_sen, 3)
print(classification_report(y_true=np.array(Teste_Y_sen), y_pred=votings, digits=3, ))
for i in range (len(Teste_Y_sen[0])+1):
    print(str(i)+"/"+str(len(Teste_Y_sen[0]))+" classes:", acerto_absoluto(Teste_Y_sen, votings, i))
```
