```python
import sqlite3
import pandas as pd

import sys
sys.path.append("../../")
from controller import *
```

```python
# Gets some more data from the database to enrich analysis.
conn = sqlite3.connect('../../radar_db')
c = conn.cursor()
c.execute("SELECT cod_ident_ato, ementa, chefe_de_governo, link_texto_original, referenda, links, num_ato FROM Decretos")
decretos = c.fetchall()
conn.close()

decretos_df = pd.DataFrame(decretos)
decretos_df.set_index(0, inplace = True)
```

## Referenda

```python
decretos_minist = load_ministerio_decrees('../../radar_db')

classes_minist=get_classes("classificacao referenda")

decretos_classify_minist = multilabel_classify_decrees(decretos_minist, classes_minist)
get_df_ministerio(decretos_classify_minist)
```

```python
print("Others referendum:")
print("\n---------------------------------------------------------------------------------------------\n")
presidentes = pd.DataFrame()
for d in decretos_classify_minist:
    if(d[2][-1] == 1):
        print("Decreto:", d[0])
        print("Link:",  decretos_df.loc[d[0]][3])
        print("Presidente:", decretos_df.loc[d[0]][2])
        try:
            presidentes[decretos_df.loc[d[0]][2]].iloc[0] += 1
        except:
            presidentes[decretos_df.loc[d[0]][2]] = [1]
        print("Referenda:", decretos_df.loc[d[0]][4])
        print("\nEmenta:\n",decretos_df.loc[d[0]][1])
        print("\n---------------------------------------------------------------------------------------------\n")
        
presidentes = presidentes.transpose()
presidentes.columns = ["Number of unknown referendum"]
print(presidentes.head())
```

## Senate

```python
decretos_df[6] = [int(n_ato.replace(".", "")) for n_ato in decretos_df[6]]
decretos_df.set_index(6, inplace=True)
```

```python
decretos_sen = load_senado_decrees('../../Senado API files/', '../../radar_db')

classes_sen=get_classes("classificacao senado")

decretos_classify_sen = multilabel_classify_decrees(decretos_sen, classes_sen)
get_df_senado(decretos_classify_sen)
```

```python
base_dir = "../../Senado API files/"
years = [i for i in range(2000,2020)]
dataframes = []
for year in years:
    direct = base_dir+str(year)+".csv"
    dataframes.append(pd.read_csv(direct, sep=';'))

dataframes = pd.concat(dataframes)
dataframes
categoria_serie = dataframes['categoria']

i=0
for row in categoria_serie:
    row = row.strip('[').strip(']').split(", ")
    row = [ro.strip('\'') for ro in row]
    dataframes.iat[i, 1] = row
    i+=1
dataframes.set_index('Numero', inplace=True)

dataframes
```

```python
presidentes = pd.DataFrame()
topics = pd.DataFrame()
for d in decretos_classify_sen:
    if(d[2][-1] == 1):
        print("Decreto:", d[0])
        print("Link:",  decretos_df.loc[d[0]][3])
        print("Presidente:", decretos_df.loc[d[0]][2])
        try:
            presidentes[decretos_df.loc[d[0]][2]].iloc[0] += 1
        except:
            presidentes[decretos_df.loc[d[0]][2]] = [1]
            
        try:
            print("Classificação Senado: ", end="")
            for i in(dataframes.loc[d[0]]['categoria']):
                print(i, end=' | ')
                try:
                    topics[i].iloc[0] += 1
                except:
                    topics[i] = [1]
        except:
            print("Não classificado")
        print("\nEmenta:\n",decretos_df.loc[d[0]][1])
        print("\n---------------------------------------------------------------------------------------------\n")
presidentes = presidentes.transpose()
presidentes.columns = ["Number of decrees that are not classified by the Senate"]
print(presidentes.head())
print("\n---------------------------------------------------------------------------------------------\n")
print("Topics:")

print(topics.transpose())
```
