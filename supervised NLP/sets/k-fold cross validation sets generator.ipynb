### Tratamento inicial dos documentos
Observações:
 * Na enumeração de artigos de um decreto são encontrados diferentes padrões. É interessante que cada artigo seja ao menos separado em uma frase. Até o nono artigo de um decreto, segue-se o "padrão art. nº". A partir do décimo, segue-se o padrão "art. 10.", que introduz um ponto final após o número. Por ser a estrutura de maior importância dos decretos, isso foi observado e tratado aqui.
 * Outra observação é referente a leis e decretos de numeração superior a 1000. A partir daí, pode ser observado o . no meio do número (ex: 1.000), que atrapalha o casamento de padrões de pontuação. Nesse caso, o ponto foi removido aqui.
 * Por fim, a assinatura do presidente e ministro responsáveis pelo decreto foi também removida. Isso evita que os algoritmos (principalmente classificadores baseados em referenda) sejam influenciados pelos ministros responsáveis por referendar o texto.

```python
from nltk.tokenize import sent_tokenize, word_tokenize

number_of_decrees = len(decretos_df['full_text'])
breaking_punctuation = ['.', '!', '?', ';', ':']
for j,decreto in enumerate(decretos_df['full_text']):
    print("Preprocessing", j+1, "/", number_of_decrees)
    text = ""
    
    decreto = decreto.replace("/", " ")
    decreto = decreto.replace("º", '')
    tokenized_document = word_tokenize(decreto)
    for i, word in enumerate(tokenized_document):
        
        if word.lower() == "art":               # lidando com os padrões de numeração de artigos definidos na observação.
            if tokenized_document[i+2].isnumeric() and tokenized_document[i+3] == ".":
                tokenized_document[i+3] = ' '
    
        if(word.isalpha() or word.isnumeric()):
            text+=(word.lower()+" ")
        elif word in breaking_punctuation:
            text+=(" . ")                       # padroniza pontuações específicas em ponto final.
        elif word[0].isnumeric() and ("," not in word):
            word = word.replace(".", "")
            if(word.isnumeric()):
                text+=(word+" ")                # lidando com números com pontuação no meio. Desconsidera números reais, comumente referentes à moeda.
    
    text = text.replace("art  .", " art ")
    text = text.replace("arts  .", " arts ")
    
    # remoção de assinaturas
    if(text.rfind(" república") != -1):
        signature = text.rfind(" república")
        if signature < len(text)/2:
            signature = len(text)
    
    decretos_df['full_text'].iloc[j] = text[:signature] + " república."
```

### 5 folds with 20% data. One will be holded out and used as test. 4-fold cross validation will be used with the others
Spliting multi-label data in a stratified way is not trivial due to class interdependence. Therefore, we use iterative stratification to get the most balanced sets.

```python
from skmultilearn.model_selection import IterativeStratification, iterative_train_test_split

X = decretos_df.values[:,:2]
Y = np.array([np.array(d) for d in decretos_df["Senado + referenda"]])

np.random.seed(44)

k_folder = IterativeStratification(n_splits=5, order = 3)
folds = k_folder.split(np.array(X), np.array(Y))

final_folds = []
for i in folds:
    final_folds.append(i[1])
    
dist = []
for i in range(5):
    dist.append(list(np.array(Y)[final_folds[i]].sum(axis=0)))
print("Distribuição final das classes entre os sets:")
pd.DataFrame(np.array(dist).T)
```

```python
for i, f in enumerate(final_folds):
    n_classes_sen = len(decretos_classify_sen[0][2]) - 1
    fold = pd.DataFrame(X[f])
    fold["Senado"] = list(Y[f][:,:n_classes_sen])
    fold["Referenda"] = list(Y[f][:,n_classes_sen:])
    fold.columns = ["Numero ato", "Ementa + texto completo", "Senado", "Referenda"]
    fold.to_csv("sets/Set "+str(i)+".csv", index=False)
```
